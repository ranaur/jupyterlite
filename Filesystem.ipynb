```python
def cat(file : str) -> None:
  import sys, os
  try:
    with open(file, 'r') as _file:
      for _line in iter((_file.readline),""):
        print(_line)
  except IOError as _err:
    sys.stderr.write("cat: " + file +
    ": " + _err.strerror + "\n")
def ls(dir : str = '.') -> None:
    import os
    print(os.listdir(dir))
def cd(dir : str = None) -> None:
    import os
    if dir is None:
        print(os.getcwd())
    else:
        os.chdir(dir)
    
```
